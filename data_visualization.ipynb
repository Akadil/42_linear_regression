# Import Required Libraries
Import pandas for data handling and matplotlib for plotting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Load the CSV Data
Use pandas to read the data.csv file and display the first few rows to verify the data structure.

```python
# Load the data from CSV file
df = pd.read_csv('data.csv')

# Display the first few rows to verify the data structure
print("Data shape:", df.shape)
print("\nFirst 10 rows:")
print(df.head(10))
```

# Create the Scatter Plot
Create a scatter plot using matplotlib with price on the x-axis and km on the y-axis.

```python
# Create scatter plot with price on x-axis and km on y-axis
plt.figure(figsize=(10, 6))
plt.scatter(df['price'], df['km'], alpha=0.7, color='blue', s=50)

# Basic plot without customization first
plt.xlabel('Price')
plt.ylabel('Kilometers')
plt.title('Car Data: Kilometers vs Price')
plt.show()
```

# Customize the Plot
Add axis labels, title, grid, and styling to make the plot more readable and professional.

```python
# Create a more customized scatter plot
plt.figure(figsize=(12, 8))

# Create the scatter plot
plt.scatter(df['price'], df['km'], alpha=0.7, color='darkblue', s=60, edgecolors='black', linewidth=0.5)

# Customize the plot
plt.xlabel('Price', fontsize=14, fontweight='bold')
plt.ylabel('Kilometers', fontsize=14, fontweight='bold')
plt.title('Car Data: Kilometers vs Price\n(Each point represents a car)', fontsize=16, fontweight='bold', pad=20)

# Add grid for better readability
plt.grid(True, alpha=0.3, linestyle='--')

# Format axes
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add some statistics as text
correlation = df['price'].corr(df['km'])
plt.text(0.05, 0.95, f'Correlation: {correlation:.3f}', transform=plt.gca().transAxes, 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8), fontsize=10)

# Adjust layout and show
plt.tight_layout()
plt.show()

# Print some insights
print(f"Number of data points: {len(df)}")
print(f"Price range: ${df['price'].min():,} - ${df['price'].max():,}")
print(f"Kilometers range: {df['km'].min():,} - {df['km'].max():,} km")
print(f"Correlation between price and km: {correlation:.3f}")
```
