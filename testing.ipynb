# Create the Scatter Plot
Create a scatter plot using matplotlib with price on the x-axis and km on the y-axis.

```python
# Create scatter plot with price on x-axis and km on y-axis
plt.figure(figsize=(10, 6))
plt.scatter(df['price'], df['km'], alpha=0.7, color='blue', s=50)

# Basic plot without customization first
plt.xlabel('Price')
plt.ylabel('Kilometers')
plt.title('Car Data: Kilometers vs Price')
plt.show()
```
